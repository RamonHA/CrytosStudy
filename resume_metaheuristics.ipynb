```python
import pandas as pd
import json
import os
```

```python
pwd = "results/metaheuristics/"
```

```python
data = {}
for i, v in enumerate(os.listdir( pwd )):
    n = 3 if i == 0 else 4
    with open( pwd + v , "r") as fp:
        data[ v.split(".")[n] ] = json.load(fp) 
```

```python
for i,v in data.items():
    print( i, v["acc"])
```

```python
for i,v in data.items():
    print(i, end = "\t")
    if isinstance( v["param"][0], list):
        print(v["param"][0])

    else:
        print(v["param"])
```
